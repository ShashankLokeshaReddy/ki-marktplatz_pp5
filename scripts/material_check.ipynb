```python
import pandas as pd
from datetime import timedelta
```

```python
df = pd.read_excel("../data/Auftragsfolgen-20211207.xlsx")
```

```python
my_list = ["ArtNr_Teil","AKNR", "MaschNr", 'Bestell_Nutzen', "LTermin","SchrittNr","ID_Maschstatus","Maschstatus",'Lieferdatum_Rohmaterial']
```

```python
short_df = df.iloc[:5000][my_list]
```

```python
# For testing a date between 2016-02-03 and 2022-01-25
today = pd.to_datetime("2016-02-16").date()
```

```python
short_df=short_df[["AKNR","Lieferdatum_Rohmaterial"]].drop_duplicates()
```

```python
def material_check(set_date):

    def bool_material_check(LT):
        # wenn Liefertermin nicht bekannt, dann False
        if pd.isnull(LT): 
            return False
        # wenn der morgige Tag schon weiter in der Zukunft liegt, als das Datum des Liefertermins, dann True
        elif (set_date+timedelta(days=1))>=LT: 
            return True
        else:
            return False

    return bool_material_check
```

```python
"""
In: Date for delivery of raw material
Out: True or False
"""
def material_check(LT):
    # wenn Liefertermin nicht bekannt, dann False
    if pd.isnull(LT): 
        return False
    # wenn der morgige Tag schon weiter in der Zukunft liegt, als das Datum des Liefertermins, dann True
    elif (today+timedelta(days=1))>=LT: 
        return True
    else:
        return False
```

```python
short_df["Material_Check"] = short_df["Lieferdatum_Rohmaterial"].apply(material_check(today))
```
